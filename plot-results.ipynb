```python
### Plot
import matplotlib.pyplot as plt


def plot_loss_acc(which, version, plot=False):
    metrics = pd.read_csv(f"logs/{which}/version_{version}/metrics.csv")

    aggreg_metrics = []
    agg_col = "epoch"
    for i, dfg in metrics.groupby(agg_col):
        agg = dict(dfg.mean())
        agg[agg_col] = i
        aggreg_metrics.append(agg)

    df_metrics = pd.DataFrame(aggreg_metrics)
    df_metrics[["train_loss", "val_loss"]].plot(
        grid=True, legend=True, xlabel="Epoch", ylabel="Loss"
    )
    plt.ylim([0., .6])
    if plot:
        
        plt.savefig(f"{which}-{version}-loss.pdf")

    df_metrics[["train_acc", "val_acc"]].plot(
        grid=True, legend=True, xlabel="Epoch", ylabel="ACC"
    )

    plt.ylim([.6, 1.])
    if plot:
        plt.savefig(f"{which}-{version}-acc.pdf")

    plt.show()
```

```python
plot_loss_acc(which="lion", version=6, plot=True) #  default params
```

```python
plot_loss_acc(which="adam", version=0, plot=True)
```

```python
plot_loss_acc(which="adamW", version=0, plot=True)
```

```python
plot_loss_acc(which="sgd-cosine", version=0, plot=True)
```
